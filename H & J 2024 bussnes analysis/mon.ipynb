```python
import pandas as pd
import numpy as np
```

```python
# Read the Excel file, skipping the first row and making the second row the header 
df = pd.read_excel("H^0J Scents -Done 33.xlsx", header=None, skiprows=2) 
# Set the first row as the header 
df.columns = df.iloc[0] 
# Drop the first row from the DataFrame 
df = df[1:]
#  # Display the DataFrame 
hj=df.copy()
```

```python
# CREATE A profit per product collumn
hj["Profit per product"]=hj["Selling Price"] - hj["Original Price per Product in Kenyan Shillings"]
hj
```

```python
# separate shoshos expenses 
hjcucu=hj[hj["Products Sold"].str.contains("Cucu",case=False,na=False)].reset_index(drop=True)
hj = hj[~hj["Products Sold"].str.contains("Cucu", case=False, na=False)].reset_index(drop=True)
```

```python
# separate all the products according to the cattegory
hjcream=hj[hj["BATH AND BODY WORKS PRODUCTS"].str.contains("Cream",case=False,na=False)].reset_index(drop=True)
hjsplash=hj[hj["BATH AND BODY WORKS PRODUCTS"].str.contains("Splash",case=False,na=False)].reset_index(drop=True)
hjlotion=hj[hj["BATH AND BODY WORKS PRODUCTS"].str.contains("Lotion",case=False,na=False)].reset_index(drop=True)
hjwash=hj[hj["BATH AND BODY WORKS PRODUCTS"].str.contains("Wash",case=False,na=False)].reset_index(drop=True)
hjmist=hj[hj["BATH AND BODY WORKS PRODUCTS"].str.contains("Mist",case=False,na=False)].reset_index(drop=True)
hjsoap=hj[hj["BATH AND BODY WORKS PRODUCTS"].str.contains("Soap",case=False,na=False)].reset_index(drop=True)
hjspray=hj[hj["BATH AND BODY WORKS PRODUCTS"].str.contains("Spray",case=False,na=False)].reset_index(drop=True)
hjdeodorant=hj[hj["BATH AND BODY WORKS PRODUCTS"].str.contains("Deodorant",case=False,na=False)].reset_index(drop=True)
hjbutter=hj[hj["BATH AND BODY WORKS PRODUCTS"].str.contains("Butter",case=False,na=False)].reset_index(drop=True)
```

```python
# first we are going to plot a combined line graph of org price and selling
import pandas as pd 
import matplotlib.pyplot as plt
#Plotting the combined line graph 
plt.figure(figsize=(10, 6)) 
plt.plot(hjcream["BATH AND BODY WORKS PRODUCTS"], hjcream["Original Price per Product in Kenyan Shillings"], marker='o', label='Original Price') 
plt.plot(hjcream["BATH AND BODY WORKS PRODUCTS"], hjcream["Selling Price"], marker='o', label='Selling Price') 
# Adding title and labels 
plt.title("Comparison Between the Original and Selling Price in Body Cream Products") 
plt.xlabel("BATH AND BODY WORKS PRODUCTS") 
plt.ylabel("Price in Kenyan Shillings (UDS / 127)") 
plt.legend() 
# Set the Y-axis limits 
plt.ylim(600, 1800) 
# Display the graph 
plt.xticks(rotation=45) 
plt.tight_layout()
plt.grid() 
plt.show()
```
